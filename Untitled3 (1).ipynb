```python
#  TASK 5: Exploratory Data Analysis (EDA)  Titanic Dataset

# =======================
# STEP 1: IMPORT LIBRARIES
# =======================
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Inline plotting
%matplotlib inline

# Set style
sns.set(style="whitegrid")

# =======================
# STEP 2: LOAD DATASET
# =======================
# Replace with your correct path to train.csv
df = pd.read_csv(r"C:\Users\train.csv")

print("First 5 rows of the dataset:")
display(df.head())

# =======================
# STEP 3: BASIC EXPLORATION
# =======================
print("\nDataset Info:")
df.info()

print("\nStatistical Summary:")
print(df.describe(include='all'))

print("\nMissing Values:")
print(df.isnull().sum())

print("\nTarget Variable - Survival Count:")
print(df['Survived'].value_counts())

# =======================
# STEP 4: DATA CLEANING
# =======================
# Fill missing values
df['Age'].fillna(df['Age'].median(), inplace=True)
df['Embarked'].fillna(df['Embarked'].mode()[0], inplace=True)

# Drop 'Cabin' due to high missingness
df.drop(columns=['Cabin'], inplace=True)

# Drop less useful columns
df.drop(columns=['PassengerId', 'Name', 'Ticket'], inplace=True)

print("\nMissing Values After Cleaning:")
print(df.isnull().sum())

# =======================
# STEP 5: UNIVARIATE ANALYSIS
# =======================
# Survival Count
sns.countplot(x='Survived', data=df)
plt.title("Survival Distribution")
plt.xlabel("Survived (0 = No, 1 = Yes)")
plt.ylabel("Count")
plt.show()

# Gender Count
sns.countplot(x='Sex', data=df)
plt.title("Gender Distribution")
plt.show()

# Pclass Count
sns.countplot(x='Pclass', data=df)
plt.title("Passenger Class Distribution")
plt.show()

# Age Distribution
df['Age'].hist(bins=30)
plt.title("Age Distribution")
plt.xlabel("Age")
plt.ylabel("Count")
plt.show()

# Fare Distribution
df['Fare'].hist(bins=30)
plt.title("Fare Distribution")
plt.xlabel("Fare")
plt.ylabel("Count")
plt.show()

# Embarked Distribution
sns.countplot(x='Embarked', data=df)
plt.title("Port of Embarkation")
plt.show()

# =======================
# STEP 6: BIVARIATE ANALYSIS
# =======================
# Survival by Gender
sns.countplot(x='Survived', hue='Sex', data=df)
plt.title("Survival by Gender")
plt.show()

# Survival by Class
sns.countplot(x='Survived', hue='Pclass', data=df)
plt.title("Survival by Passenger Class")
plt.show()

# Survival by Embarked
sns.countplot(x='Survived', hue='Embarked', data=df)
plt.title("Survival by Embarkation Port")
plt.show()

# Age vs Survival (Boxplot)
sns.boxplot(x='Survived', y='Age', data=df)
plt.title("Age vs Survival")
plt.show()

# Fare vs Survival (Boxplot)
sns.boxplot(x='Survived', y='Fare', data=df)
plt.title("Fare vs Survival")
plt.show()

# =======================
# STEP 7: CORRELATION ANALYSIS
# =======================
# Encode categorical columns
df_encoded = pd.get_dummies(df, drop_first=True)

# Correlation Matrix
plt.figure(figsize=(10, 6))
sns.heatmap(df_encoded.corr(), annot=True, cmap="coolwarm")
plt.title("Correlation Heatmap")
plt.show()

# Pairplot (subset for clarity)
sns.pairplot(df_encoded[['Survived', 'Age', 'Fare', 'Pclass', 'Sex_male']])
plt.suptitle("Pairplot of Selected Features", y=1.02)
plt.show()

# =======================
# STEP 8: OBSERVATIONS
# =======================
# You can write your observations in markdown cells in the notebook after each plot.

# =======================
# STEP 9: FINAL INSIGHTS SUMMARY (To be added in markdown)
# =======================
# Example summary in a markdown cell:
# - Women had a higher survival rate than men.
# - 1st Class passengers had better survival chances.
# - Most passengers embarked from port S.
# - Younger passengers and those who paid higher fares had better survival chances.
```
